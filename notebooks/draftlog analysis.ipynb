```python
event_dirs = [
    '/mnt/c/projects/ocl/2019',
    '/mnt/c/projects/ocl'
]

draft_logs = []

for event_dir in event_dirs:
    sub_dirs = os.listdir(event_dir)
    for sub_dir in sub_dirs:
        if os.path.isdir(os.path.join(event_dir, sub_dir)):
            logs = [f for f in os.listdir(os.path.join(event_dir, sub_dir)) if 'draftlog' in f]
            if len(logs) == 8:
                draft_logs.append({
                    'event_name': sub_dir,
                    'event_path': os.path.join(event_dir, sub_dir),
                    'log_files': logs
                })
            elif len(logs) > 0:
                print(f'event {sub_dir} missing logs')

pprint(draft_logs)
            
```
