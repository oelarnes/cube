```python
df = pd.DataFrame(r.json()['data']['cubeByType']['cards']).sort_values('bayesianWinRate', ascending=False)
format_dict = {
    'matchWinsInPool': '{:.0f}', 
    'matchLossesInPool': '{:.0f}', 
    'bayesianWinRate': '{:.2%}',
    'bayesianWinRate': '{:.2%}',
    'mainDeckPct': '{:.2%}', 
    'avgPickOrder': '{:.1f}'
}
df.head(25).style.format(format_dict).hide_index()
```
